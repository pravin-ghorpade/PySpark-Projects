```
# Taking the data from csv file and putting in dataframe
fb = spark.read.option("header","true").csv("/FileStore/tables/train.csv")
```

```
fb.display()
```

```
# Putting the two columns into two different arrays
src = []
dest = []
for i in fb.collect():
  x=i
  src.append(x.__getattr__("source_node"))
  dest.append(x.__getattr__("destination_node"))
```

```
# Putting all the friends of the left side element into a dictionary
fb_dict = {}
for i in src:
  fb_dict[i] = []
 
for i in range(len(src)):
  fb_dict[src[i]].append(dest[i])
```

```
# The count of all the unique possible friends
unique_ppl = len(fb_dict)
unique_ppl
```

```
# The dictionary where key is the person and the values contain all the direct friends of the key
fb_dict
```

```
# The function that calculates the probability that two people are friends
def func(test1, test2):
  prob = 1
  arr = [test1]
  val = arr[0]
  x =0
  for item in arr:
    c = 0
    if item in fb_dict:
      for i in fb_dict[item]:
        if i == test2:
          x = 1
          break
        if i not in arr:
          arr.append(i)
          prob*=(1.0/len(fb_dict[item]))
          c=c+1
    if x==1:
      break
  if x == 0:
    prob = 0
  return prob   
```

```
# Reading the csv file that contains list of two random people within the limit
test = spark.read.option("header","true").csv("/FileStore/tables/friend_test_data.csv")
```

```
# Displaying the above dataframe
test.display()
```

```
# Putting the columns of the above dataframe into two arrays
friend1 = []
friend2 = []
for i in test.collect():
  x=i
  friend1.append(x.__getattr__("friend1"))
  friend2.append(x.__getattr__("friend2"))
```

```
# The number of test cases taken
len(friend1)
```

```
# Creating a dictionary with the respective probabilities
test_dict = {}
for i in range(len(friend1)):
  test_dict[(friend1[i],friend2[i])] = func(friend1[i],friend2[i])
```

```
# The dictionary where the key is the two people and value is the probability of them being friends
test_dict
```

```
# Creating a list of the final values
prob_list = []
for key in test_dict:
  prob_list.append((key[0],key[1],str(test_dict[key])))
```

```
# The above mentioned array
prob_list
```

```
# Storing the data into a dataframe
output_df = spark.createDataFrame(data=prob_list, schema = ["Friend 1","Friend 2","Probability"])
```

```
# Displaying the dataframe
output_df.display()
```

```
# Storing the dataframe into a csv file in the hdfs
output_df.write.option("header","true").csv("dbfs:/proboutput.csv")
```
